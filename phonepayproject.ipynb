```python
query='''with quaterly_total as (
    select "State", "Year", "Quater", sum(payment_amount) as total_payment
    from agg_trans
    group by "State", "Year", "Quater"
),
ranked_quater as (
    select
        "State",
        "Year",
        "Quater",
        total_payment, -- This was 'total_amount' in your error message, corrected to 'total_payment'
        row_number() over (partition by "Year" order by total_payment desc) as rn
    from quaterly_total
)
select "State", "Year", "Quater", total_payment -- This was 'total_amount', corrected to 'total_payment'
from ranked_quater
where rn = 1;;'''
quater=pd.read_sql(query,engine)
quater

```
